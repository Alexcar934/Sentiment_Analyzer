```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/MyReviews.csv',index_col=0)

i = 1
while i < 10:
    estrellas = 1
    resena = str(input('Reseña'))

    nueva_fila = {'Estrellas':estrellas, 'Reseñas':resena}

    # Crear un nuevo dataframe con la nueva fila
    nuevo_df = pd.DataFrame(nueva_fila, index=[0])

    # Concatenar el nuevo dataframe con el dataframe existente
    df = pd.concat([df, nuevo_df], ignore_index=True)
    
    i+=1
```

```python
df.to_csv('../data/raw/MyReviews.csv')
```
