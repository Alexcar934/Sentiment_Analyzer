```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import time
import numpy as np
```

### Reviews Amazon

```python
url  = 'https://es.trustpilot.com/review/amazon.es'

response = requests.get(url)
soup = bs(response.content, 'lxml')

soup
```

```python

lista_estrellas = []
lista_reseñas = []

num_paginas = int(soup.find('a', attrs={'name':'pagination-button-last'}).get_text())
print(num_paginas)

for pagina in range(num_paginas+1):
    print('Página: ',str(pagina))
    response = requests.get(url + '?page=' + str(pagina))
    soup_pagina = bs(response.content, 'lxml')

    contenedor = soup_pagina.find_all('section', class_='styles_reviewContentwrapper__zH_9M')

    for i in contenedor:
        lista_estrellas.append(i.find('div',class_='styles_reviewHeader__iU9Px').get('data-service-review-rating'))
        lista_reseñas.append(i.find('div',class_='styles_reviewContent__0Q2Tg').get_text().split('Fecha de la experiencia')[0])

len(lista_reseñas)

```

```python
df = pd.DataFrame()

df['Estrellas'] = lista_estrellas
df['Reseñas'] = lista_reseñas
```

```python
df.to_csv('../data/raw/Reviews_amazon.csv')
```
